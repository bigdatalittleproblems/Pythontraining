In the example below, we have pandas use the “read_csv” method to read directly from the “raw” github repo, and assign the data frame to the variable “df”. We could have selected any value to assign the dataframe to. the we use the "head(20)" method to see the top 20 values.

```
df=pd.read_csv('https://raw.githubusercontent.com/datadesk/california-coronavirus-data/master/latimes-county-totals.csv')

df.head(20)
```

We use the pivot method to reshape the data and look at "new_confirmed_cases" by county over time. we assign the output to the "df_pivot" variable then use the tail() Method to look at the last 20 values of the data frame. You can change any of the parameters to look at different cuts of data.

```
df_pivot=df.pivot(index='date',columns='county',values='new_confirmed_cases')
# use tail to see the last values that are added.
df_pivot.tail(20)
```

This is the simplest plot method, but without selecting columns it will give you every county in CA, and this can be overwhelming to look at. 

```
df_pivot.plot()
```

we use the columns property to see what are the avaiable columns to select.



```
df_pivot.columns
```

Here we select the columns that we want to look at and then pass it through the plot method. 

```
col=['Los Angeles','Orange','San Bernardino','San Francisco']
df_pivot[col].plot()
```
